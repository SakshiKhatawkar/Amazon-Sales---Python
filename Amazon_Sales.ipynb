# EDA

```python
# 1.Find average discounted and original price.\
df[['discounted_price','original_price']].mean()
```

```python
# 2.Top 10 products with highest discount percentage.
df.sort_values('discount_percentage', ascending=False)[['product_title','discount_percentage']].head(10)
```

```python
# 3.Analyze the distribution of discounted product prices.
df['discounted_price'].plot(kind='hist', bins=30, title='Distribution of Discounted Prices')
```

```python
# 4.Compare average product ratings across different categories
df.groupby('product_category')['product_rating'].mean().plot(kind='bar', title='Average Rating by Category')
```

```python
# 5.Identify the top 10 product categories based on monthly purchases.
df.groupby('product_category')['purchased_last_month'].sum() \
  .sort_values(ascending=False).head(10) \
  .plot(kind='bar', title='Top 10 Categories by Purchases')
```

```python
# 6.Average rating per product category.
df.groupby('product_category')['product_rating'].mean().sort_values(ascending=False)
```

```python
# 7.Products with rating > 4.5 and reviews > 5000.
df[(df['product_rating'] > 4.5) & (df['total_reviews'] > 5000)]
```

```python
# 8.Correlation between rating and reviews.
df[['product_rating','total_reviews']].corr()
```

```python
# 9. Is there any relationship between product price and customer rating?
df.plot(kind='scatter',x='discounted_price',y='product_rating',title='Price vs Rating')
```

```python
# 10.Analyze how discount percentages are distributed.
df['discount_percentage'].plot(kind='hist', bins=25, title='Discount Percentage Distribution')
```

```python
# 11. Compare average sales of sponsored vs non-sponsored products.
df.groupby('is_sponsored')['purchased_last_month'].mean()
```

```python
# 12.Best-seller vs non-best-seller average rating
df.groupby('is_best_seller')['product_rating'].mean()
```

```python
# 13.Which category has highest monthly purchases?
df.groupby('product_category')['purchased_last_month'].sum().sort_values(ascending=False)
```

```python
# 14.Does sponsorship affect monthly purchases?
df.groupby('is_sponsored')['purchased_last_month'].mean().plot(kind='bar', title='Sponsored vs Non-Sponsored Sales')
```

```python
# 15.Analyze the price difference between original and discounted prices.
(df['original_price'] - df['discounted_price']).plot(kind='hist', bins=40, title='Price Difference Distribution')
```

```python
# 16.Does higher discount lead to higher sales?
df.plot(kind='scatter',x='discount_percentage',y='purchased_last_month',title='Discount vs Monthly Purchases')
```

```python
# 17. Analyze price variability and outliers across different product categories.
df.boxplot(column='discounted_price', by='product_category', figsize=(12,6))
```

```python
# 18.Analyze the proportion of sponsored and non-sponsored products.
df['is_sponsored'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Sponsored vs Non-Sponsored')
```

```python
# 19. Compare rating distributions between best-seller and non-best-seller products.
df.violinplot = df.boxplot(column='product_rating', by='is_best_seller')
```

```python
# 20.Visualize coupon vs non-coupon product distribution.
sizes = df['has_coupon'].value_counts()
plt.pie(sizes,autopct='%1.1f%%')
plt.title('Coupon Availability')
plt.show()
```
